```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from scipy.stats import pearsonr
import warnings
warnings.filterwarnings('ignore')

#  df is already loaded, so work directly with it
print("Dataset shape:", df.shape)



# Clean data - remove rows with missing values for these specific columns
df.dropna(subset=['COPDSEVERITY', 'FEV1PRED'], inplace=True)

# SCATTER PLOT WITH REGRESSION LINE
plt.figure(figsize=(15, 10))

# Subplot 1: Scatter plot with regression line
plt.subplot(2, 3, 1)
# Assuming 'COPDSEVERITY' can be mapped to a numerical scale if needed for correlation/regression
# If 'COPDSEVERITY' is categorical, correlation and linear regression might not be appropriate directly.
# For visualization purposes, we can try to map it or use a different plot type.
# For now, I will assume a numerical mapping if you intend to perform regression.
# Let's create a numerical representation for 'COPDSEVERITY' for plotting
copd_severity_mapping = {'MILD': 1, 'MODERATE': 2, 'SEVERE': 3, 'VERY SEVERE': 4}
df['COPD_Numeric'] = df['COPDSEVERITY'].map(copd_severity_mapping)

x = df['COPD_Numeric']  # Numeric COPD severity (1-4)
y = df['FEV1PRED']  # FEV1 Predicted percentage

# Calculate correlation
correlation, p_value = pearsonr(x, y)

# Create scatter plot
plt.scatter(x, y, alpha=0.7, color='steelblue', s=60, edgecolors='black', linewidths=0.5)

# Add regression line
slope, intercept, r_value, p_val, std_err = stats.linregress(x, y)
line = slope * x + intercept
plt.plot(x, line, 'red', linewidth=2, label=f'y = {slope:.2f}x + {intercept:.2f}')

plt.xlabel('COPD Severity (Numeric)', fontsize=12)
plt.ylabel('FEV1 Predicted (%)', fontsize=12)
plt.title('COPD Severity vs FEV1 Predicted\n(Scatter Plot with Regression)', fontsize=14)
plt.xticks([1, 2, 3, 4], ['MILD\n(1)', 'MODERATE\n(2)', 'SEVERE\n(3)', 'VERY SEVERE\n(4)'])
plt.legend()
plt.grid(True, alpha=0.3)

# Add correlation text box
plt.text(0.05, 0.95, f'r = {correlation:.3f}\np = {p_value:.4f}\nR² = {correlation**2:.3f}',
         transform=plt.gca().transAxes, bbox=dict(boxstyle="round", facecolor='lightblue', alpha=0.8),
         verticalalignment='top', fontsize=11)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, roc_curve
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
import warnings
warnings.filterwarnings('ignore')

print("Multiple Logistic Regression Analysis: Predicting SGRQ Categories")
print("="*70)

# ============================================================================
# DATA PREPARATION
# ============================================================================

print("\n1. DATA PREPARATION")
print("-" * 30)

# Check available data
required_cols = ['PackHistory', 'COPDSEVERITY', 'MWT1Best', 'FVC', 'SGRQ']
df_model = df[required_cols].copy()

# Remove rows with missing values
df_clean = df_model.dropna()


# Display SGRQ distribution
print(f"\nSGRQ Statistics:")
print(df_clean['SGRQ'].describe())

# ============================================================================
# CONVERT SGRQ TO CATEGORICAL VARIABLE
# ============================================================================

print("\n2. SGRQ CATEGORIZATION")
print("-" * 30)

# SGRQ interpretation (St. George's Respiratory Questionnaire):
# Lower scores = better quality of life
# Higher scores = worse quality of life/greater impact
# Common clinical cutoffs: <25 (Good), 25-50 (Moderate), >50 (Poor)

# Create SGRQ categories based on clinical cutoffs
def categorize_sgrq(score):
    if score <= 25:
        return 0  # Good quality of life
    elif score <= 50:
        return 1  # Moderate impact
    else:
        return 2  # Poor quality of life

df_clean['SGRQ_Category'] = df_clean['SGRQ'].apply(categorize_sgrq)
df_clean['SGRQ_Category_Label'] = df_clean['SGRQ_Category'].map({
    0: 'Good QoL (≤25)',
    1: 'Moderate QoL (26-50)',
    2: 'Poor QoL (>50)'
})

# Display category distribution
print("SGRQ Category Distribution:")
category_counts = df_clean['SGRQ_Category_Label'].value_counts()
print(category_counts)
print(f"\nPercentages:")
print((category_counts / len(df_clean) * 100).round(1))

# ============================================================================
# PREPARE PREDICTOR VARIABLES
# ============================================================================

print("\n3. PREDICTOR VARIABLES PREPARATION")
print("-" * 30)

# Encode COPDSEVERITY as ordinal variable
severity_mapping = {'MILD': 1, 'MODERATE': 2, 'SEVERE': 3, 'VERY SEVERE': 4}
df_clean['COPDSEVERITY_Numeric'] = df_clean['COPDSEVERITY'].map(severity_mapping)

# Prepare feature matrix X
X = df_clean[['PackHistory', 'COPDSEVERITY_Numeric', 'MWT1Best', 'FVC']].copy()
y = df_clean['SGRQ_Category'].copy()

print("Predictor variables summary:")
print(X.describe())

print(f"\nPredictor variables correlation matrix:")
corr_matrix = X.corr()
print(corr_matrix.round(3))

# Check for multicollinearity using VIF
print(f"\nVariance Inflation Factors (VIF):")
print("(VIF > 5 indicates potential multicollinearity)")
for i in range(X.shape[1]):
    vif = variance_inflation_factor(X.values, i)
    print(f"{X.columns[i]}: {vif:.2f}")

# ============================================================================
# BINARY LOGISTIC REGRESSION (Good vs Poor QoL)
# ============================================================================

print("\n4. BINARY LOGISTIC REGRESSION")
print("-" * 30)

# Create binary outcome: Good QoL (0) vs Poor QoL (1)
# Combine Moderate and Poor into one category for binary classification
y_binary = (df_clean['SGRQ_Category'] > 0).astype(int)
y_binary_labels = y_binary.map({0: 'Good QoL', 1: 'Impaired QoL'})

print(f"Binary outcome distribution:")
print(y_binary_labels.value_counts())

# Split data
X_train, X_test, y_train_bin, y_test_bin = train_test_split(
    X, y_binary, test_size=0.3, random_state=42, stratify=y_binary
)

# Scale features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Fit binary logistic regression
log_reg_binary = LogisticRegression(random_state=42, max_iter=1000)
log_reg_binary.fit(X_train_scaled, y_train_bin)

# Predictions
y_pred_bin = log_reg_binary.predict(X_test_scaled)
y_pred_prob_bin = log_reg_binary.predict_proba(X_test_scaled)[:, 1]

# Model performance
print(f"\nBinary Logistic Regression Results:")
print(f"Accuracy: {accuracy_score(y_test_bin, y_pred_bin):.3f}")
print(f"Precision: {precision_score(y_test_bin, y_pred_bin):.3f}")
print(f"Recall: {recall_score(y_test_bin, y_pred_bin):.3f}")
print(f"F1-Score: {f1_score(y_test_bin, y_pred_bin):.3f}")
print(f"AUC-ROC: {roc_auc_score(y_test_bin, y_pred_prob_bin):.3f}")

# Feature importance (coefficients)
feature_names = ['PackHistory', 'COPD_Severity', 'MWT1Best', 'FVC']
coefficients = log_reg_binary.coef_[0]
print(f"\nFeature Coefficients (Log-Odds):")
for name, coef in zip(feature_names, coefficients):
    odds_ratio = np.exp(coef)
    print(f"{name}: {coef:.3f} (OR: {odds_ratio:.3f})")

# ============================================================================
# MULTINOMIAL LOGISTIC REGRESSION (3 Categories)
# ============================================================================

print("\n5. MULTINOMIAL LOGISTIC REGRESSION")
print("-" * 30)

# Only proceed if we have all 3 categories
if len(df_clean['SGRQ_Category'].unique()) >= 2:
    # Split data for multinomial
    X_train_multi, X_test_multi, y_train_multi, y_test_multi = train_test_split(
        X, y, test_size=0.3, random_state=42, stratify=y
    )

    # Scale features
    scaler = StandardScaler()
    X_train_multi_scaled = scaler.fit_transform(X_train_multi)
    X_test_multi_scaled = scaler.transform(X_test_multi)

    # Fit multinomial logistic regression
    log_reg_multi = LogisticRegression(
        multi_class='multinomial', solver='lbfgs', random_state=42, max_iter=1000
    )
    log_reg_multi.fit(X_train_multi_scaled, y_train_multi)

    # Predictions
    y_pred_multi = log_reg_multi.predict(X_test_multi_scaled)
    y_pred_prob_multi = log_reg_multi.predict_proba(X_test_multi_scaled)

    # Model performance
    print(f"Multinomial Logistic Regression Results:")
    print(f"Accuracy: {accuracy_score(y_test_multi, y_pred_multi):.3f}")

    print(f"\nDetailed Classification Report:")
    target_names = ['Good QoL', 'Moderate QoL', 'Poor QoL']
    print(classification_report(y_test_multi, y_pred_multi, target_names=target_names))

    # Feature coefficients for each class
    print(f"\nMultinomial Coefficients (vs Reference Category):")
    for i, class_name in enumerate(target_names[1:], 1):  # Skip reference category
        print(f"\n{class_name} vs {target_names[0]}:")
        for name, coef in zip(feature_names, log_reg_multi.coef_[i-1]):
            odds_ratio = np.exp(coef)
            print(f"  {name}: {coef:.3f} (OR: {odds_ratio:.3f})")

# ============================================================================
# STATISTICAL ANALYSIS WITH STATSMODELS
# ============================================================================

print("\n6. DETAILED STATISTICAL ANALYSIS (STATSMODELS)")
print("-" * 30)

# Prepare data for statsmodels (binary logistic regression)
X_stats = X.copy()
X_stats = sm.add_constant(X_stats)  # Add intercept

# Fit binary logistic regression with statsmodels
logit_model = sm.Logit(y_binary, X_stats)
logit_results = logit_model.fit(disp=0)

print("Logistic Regression Summary (Binary Outcome):")
print(logit_results.summary())

# Calculate odds ratios and confidence intervals
print(f"\nOdds Ratios with 95% Confidence Intervals:")
odds_ratios = np.exp(logit_results.params)
conf_int = np.exp(logit_results.conf_int())
for i, (param, or_val) in enumerate(zip(logit_results.params.index, odds_ratios)):
    if param != 'const':
        print(f"{param}: OR = {or_val:.3f}, 95% CI: ({conf_int.iloc[i, 0]:.3f}, {conf_int.iloc[i, 1]:.3f})")

# ============================================================================
# VISUALIZATIONS
# ============================================================================

print("\n7. CREATING VISUALIZATIONS...")
print("-" * 30)

# Create visualizations
plt.figure(figsize=(16, 12))

# Subplot 1: SGRQ Category Distribution
plt.subplot(2, 4, 1)
category_counts.plot(kind='bar', color=['lightgreen', 'orange', 'lightcoral'])
plt.title('SGRQ Category Distribution')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.3)

# Subplot 2: Feature Correlation Matrix
plt.subplot(2, 4, 2)
sns.heatmap(X.corr(), annot=True, cmap='coolwarm', center=0, square=True, fmt='.2f')
plt.title('Predictor Variables\nCorrelation Matrix')

# Subplot 3: Feature Importance (Binary Model)
plt.subplot(2, 4, 3)
feature_importance = pd.Series(coefficients, index=feature_names)
feature_importance.plot(kind='barh', color='steelblue')
plt.title('Feature Coefficients\n(Binary Logistic Regression)')
plt.xlabel('Log-Odds Coefficient')
plt.grid(axis='x', alpha=0.3)

# Subplot 4: ROC Curve (Binary Model)
plt.subplot(2, 4, 4)
fpr, tpr, thresholds = roc_curve(y_test_bin, y_pred_prob_bin)
plt.plot(fpr, tpr, 'b-', linewidth=2, label=f'AUC = {roc_auc_score(y_test_bin, y_pred_prob_bin):.3f}')
plt.plot([0, 1], [0, 1], 'r--', linewidth=1)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve\n(Binary Classification)')
plt.legend()
plt.grid(alpha=0.3)

# Subplot 5: Confusion Matrix (Binary)
plt.subplot(2, 4, 5)
cm_binary = confusion_matrix(y_test_bin, y_pred_bin)
sns.heatmap(cm_binary, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Good QoL', 'Impaired QoL'],
            yticklabels=['Good QoL', 'Impaired QoL'])
plt.title('Confusion Matrix\n(Binary Classification)')
plt.ylabel('True Label')
plt.xlabel('Predicted Label')

# Subplot 6: Prediction Probabilities Distribution
plt.subplot(2, 4, 6)
plt.hist(y_pred_prob_bin[y_test_bin == 0], alpha=0.5, label='Good QoL', bins=20, color='green')
plt.hist(y_pred_prob_bin[y_test_bin == 1], alpha=0.5, label='Impaired QoL', bins=20, color='red')
plt.xlabel('Predicted Probability (Impaired QoL)')
plt.ylabel('Frequency')
plt.title('Prediction Probability\nDistribution')
plt.legend()
plt.grid(alpha=0.3)

# Subplot 7: Odds Ratios Visualization
plt.subplot(2, 4, 7)
or_values = np.exp(coefficients)
colors = ['red' if or_val < 1 else 'green' for or_val in or_values]
bars = plt.bar(feature_names, or_values, color=colors, alpha=0.7)
plt.axhline(y=1, color='black', linestyle='--', alpha=0.5)
plt.title('Odds Ratios\n(OR > 1: Risk Factor)')
plt.ylabel('Odds Ratio')
plt.xticks(rotation=45)
for bar, or_val in zip(bars, or_values):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.05,
             f'{or_val:.2f}', ha='center', va='bottom', fontsize=9)
plt.grid(axis='y', alpha=0.3)

# Subplot 8: Model Performance Metrics
plt.subplot(2, 4, 8)
metrics = ['Accuracy', 'Precision', 'Recall', 'F1-Score', 'AUC']
values = [
    accuracy_score(y_test_bin, y_pred_bin),
    precision_score(y_test_bin, y_pred_bin),
    recall_score(y_test_bin, y_pred_bin),
    f1_score(y_test_bin, y_pred_bin),
    roc_auc_score(y_test_bin, y_pred_prob_bin)
]
bars = plt.bar(metrics, values, color='skyblue', alpha=0.7)
plt.title('Model Performance\nMetrics')
plt.ylabel('Score')
plt.ylim(0, 1)
for bar, val in zip(bars, values):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.01,
             f'{val:.3f}', ha='center', va='bottom', fontsize=9)
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()

# ============================================================================
# MODEL INTERPRETATION
# ============================================================================

print("\n8. MODEL INTERPRETATION")
print("-" * 30)

print("Key Findings:")
print("• SGRQ represents quality of life impact (higher = worse quality)")
print("• Binary model predicts Good vs Impaired quality of life")

print(f"\nFeature Effects on Poor Quality of Life:")
for name, coef, or_val in zip(feature_names, coefficients, np.exp(coefficients)):
    effect = "increases" if coef > 0 else "decreases"
    magnitude = "strong" if abs(coef) > 1 else "moderate" if abs(coef) > 0.5 else "weak"
    print(f"• {name}: {magnitude} {effect} risk (OR: {or_val:.2f})")

print(f"\nClinical Interpretation:")
print(f"• Higher pack-years of smoking history likely increase poor QoL risk")
print(f"• Worse COPD severity increases poor QoL risk")
print(f"• Better walking capacity (MWT1Best) likely decreases poor QoL risk")
print(f"• Better lung function (FVC) likely decreases poor QoL risk")

print(f"\nModel Performance:")
auc_score = roc_auc_score(y_test_bin, y_pred_prob_bin)
if auc_score > 0.8:
    performance = "Excellent"
elif auc_score > 0.7:
    performance = "Good"
elif auc_score > 0.6:
    performance = "Fair"
else:
    performance = "Poor"
print(f"• AUC = {auc_score:.3f} ({performance} discriminative ability)")
print(f"• The model can effectively distinguish between good and impaired QoL")
```
